```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Heart Attack Data Set.csv")
df
```

```python
df.target.value_counts()
```

```python
sns.countplot(x="target", data=df)
plt.show()
```

```python
NoDisease = len(df[df.target==0])
YesDisease = len(df[df.target==1])
print("Percentage of Patients Haven't Heart Disease: {:.2f}%".format((NoDisease/(len(df.target))*100)))
print("Percentage of Patients Have Heart Disease: {:.2f}%".format((YesDisease/(len(df.target))*100)))
```

```python
countFemale = len(df[df.sex==0])
countMale = len(df[df.sex==1])
print("Percentage of Female Patients: {:.2f}%".format((countFemale/(len(df.sex))*100)))
print("Percentage of Male Patients: {:.2f}%".format((countMale/(len(df.sex))*100)))
```

```python
df.sex.value_counts()
```

```python
sns.countplot(x='sex',data=df)
plt.xlabel("Sex (0=Female, 1=Male)")
plt.show()
```

```python
pd.crosstab(df.age,df.target).plot(kind="bar",figsize=(20,6))
plt.title("Heart Disease Frequency for Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
pd.crosstab(df.sex,df.target).plot(kind="bar",figsize=(15,6))
plt.title("Heart Disease Frequency for Sex")
plt.xlabel("Sex( (0=Female, 1=Male)")
plt.ylabel("Frequency")
plt.xticks(rotation=0)
plt.show()
```

```python
pd.crosstab(df.cp,df.target).plot(kind="bar",figsize=(15,6))
plt.title('Heart Disease Frequency According To Chest Pain Type')
plt.xlabel('Chest Pain Type')
plt.xticks(rotation = 0)
plt.ylabel('Frequency of Disease or Not')
plt.show()
```

```python
plt.scatter(x=df.age[df.target==1], y=df.thalach[(df.target==1)])
plt.scatter(x=df.age[df.target==0], y=df.thalach[(df.target==0)])
plt.legend(["Not Disease", "Disease"])
plt.xlabel("Age")
plt.ylabel("Maximum Heart Rate")
plt.show()
```

```python
# Plotting Histogram
plt.figure(figsize=(10, 6))
plt.hist(df['age'],color='skyblue',edgecolor='black')
plt.title('Histogram of age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Assuming the dataset has a 'date' column for time series
plt.figure(figsize=(10, 6))
plt.plot(df['sex'], df['target'])
plt.title('disease over years')
plt.xlabel('Year')
plt.ylabel('disease')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
dg = pd.read_csv("dataset_Facebook.csv",sep=';')
dg
```

```python
# Assuming the dataset has a 'date' column for time series
plt.figure(figsize=(10, 6))
plt.plot(dg['Post Month'], dg['Type'])
plt.title('disease over years')
plt.xlabel('Year')
plt.ylabel('disease')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
