```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('dataset_Facebook.csv',sep=";")
df
```

```python
plt.title("BOXPLOT")
sns.boxplot(x="Post Hour",data=df)
```

```python
plt.title("HISTOGRAM")
sns.histplot(x='Lifetime Post Total Reach',bins=50,data=df)
plt.xlabel('Lifetime Post Total Reach')
plt.ylabel('Frequency')
plt.show()
```

```python
month_data = df[df['Post Month'] == 4]
# Plot the single line graph for Post Month 3 and Lifetime Post Total Reach
plt.title("Single Line Graph for Post Month 4")
plt.plot(month_data['Lifetime Post Total Reach'])
plt.xlabel('Index')
plt.ylabel('Lifetime Post Total Reach')
plt.show()
```

```python
#hue=sed to differentiate categories or groups within your data
plt.title("Multiple line graph")
sns.lineplot(data=df,x="Post Month",y="Lifetime Post Total Reach",hue="Type")
plt.xlabel('Month')
plt.ylabel('Lifetime Post Total Reach')
plt.show()
```
